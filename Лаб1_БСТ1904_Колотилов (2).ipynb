```python
from random import randint
def random_matrix(m = 50, n = 50, min_limit = -250, max_limit = 1008):
    matrix = [[randint(min_limit, max_limit) for j in range(m)] for i in range(n)]
    return matrix
print(random_matrix(3, 3, 10, 99))
```

```python
### Задание №3(Сортировка выбором)
```

```python
import time
def selection_sort(matrix):
    for row in matrix:
        for i in range(len(row)):
            min_index = i
            for j in range(i + 1, len(row)):
                if row[j] < row[min_index]:
                    min_index = j
            row[i], row[min_index] = row[min_index], row[i]
        print(row)
start = time.time()
matrix = random_matrix(3, 3, 10, 99)
print("Исходная матрица")
for row in matrix:
    print(row)
print("Матрица с отсортированными строками")
selection_sort(matrix)
print("Время работы:", time.time() - start)
```

```python
### Задание №3(Сортировка вставкой)
```

```python
def insertion_sort(matrix):  
    for row in matrix:
        for i in range(1, len(row)):
            item_to_insert = row[i]
            j = i - 1
            while j >= 0 and row[j] > item_to_insert:
                row[j + 1] = row[j]
                j -= 1
            row[j + 1] = item_to_insert
        print(row)
start = time.time()
matrix = random_matrix(3, 3, 10, 99)
print("Исходная матрица")
for row in matrix:
    print(row)
print("Матрица с отсортированными строками")
insertion_sort(matrix)
print("Время работы:", time.time() - start)
```

```python
### Задание №3(Сортировка обменом)
```

```python
def bubble_sort(matrix):  
    for row in matrix:
        swapped = True
        while swapped:
            swapped = False
            for i in range(len(row) - 1):
                if row[i] > row[i + 1]:
                    row[i], row[i + 1] = row[i + 1], row[i]
                    swapped = True
        print(row)
start = time.time()
matrix = random_matrix(3, 3, 10, 99)
print("Исходная матрица")
for row in matrix:
    print(row)
print("Матрица с отсортированными строками")
bubble_sort(matrix)
print("Время работы:", time.time() - start)
```

```python
### Задание №3(Сортировка Шелла)
```

```python
def shell_sort(matrix):
    for row in matrix:
        inc = len(row) // 2
        while inc:
            for i, el in enumerate(row):
                while i >= inc and row[i - inc] > el:
                    row[i] = row[i - inc]
                    i -= inc
                row[i] = el
            inc = 1 if inc == 2 else int(inc * 5.0 / 11)
        print(row)
start = time.time()
matrix = random_matrix(3, 3, 10, 99)
print("Исходная матрица")
for row in matrix:
    print(row)
print("Матрица с отсортированными строками")
shell_sort(matrix)
print("Время работы:", time.time() - start)
```

```python
### Задание №3(Сортировка пирамидальная)
```

```python
def heapify(nums, heap_size, root_index):  
    largest = root_index
    left_child = (2 * root_index) + 1
    right_child = (2 * root_index) + 2
    if left_child < heap_size and nums[left_child] > nums[largest]:
        largest = left_child
    if right_child < heap_size and nums[right_child] > nums[largest]:
        largest = right_child
    if largest != root_index:
        nums[root_index], nums[largest] = nums[largest], nums[root_index]
        heapify(nums, heap_size, largest)

def heap_sort(matrix):  
    for row in matrix:
        n = len(row)
        for i in range(n, -1, -1):
            heapify(row, n, i)
        for i in range(n - 1, 0, -1):
            row[i], row[0] = row[0], row[i]
            heapify(row, i, 0)
        print(row)
        
start = time.time()
matrix = random_matrix(3, 3, 10, 99)
print("Исходная матрица")
for row in matrix:
    print(row)
print("Матрица с отсортированными строками")
heap_sort(matrix)
print("Время работы:", time.time() - start)
```

```python
### Задание №3(Сортировка быстрая)
```

```python
def partition(nums, low, high):  
    # Выбираем средний элемент в качестве опорного
    # Также возможен выбор первого, последнего
    # или произвольного элементов в качестве опорного
    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while nums[i] < pivot:
            i += 1

        j -= 1
        while nums[j] > pivot:
            j -= 1

        if i >= j:
            return j

        # Если элемент с индексом i (слева от опорного) больше, чем
        # элемент с индексом j (справа от опорного), меняем их местами
        nums[i], nums[j] = nums[j], nums[i]

def quick_sort(matrix): 
    for row in matrix:
        # Создадим вспомогательную функцию, которая вызывается рекурсивно
        def _quick_sort(items, low, high):
            if low < high:
                # This is the index after the pivot, where our lists are split
                split_index = partition(items, low, high)
                _quick_sort(items, low, split_index)
                _quick_sort(items, split_index + 1, high)

        _quick_sort(row, 0, len(row) - 1)
        print(row)
start = time.time()
matrix = random_matrix(3, 3, 10, 99)
print("Исходная матрица")
for row in matrix:
    print(row)
print("Матрица с отсортированными строками")
quick_sort(matrix)
print("Время работы:", time.time() - start)
```
